### Example with random data

```python
#example on multiline plot
n_estimators = [50,100,200,300,400,500]
max_depth = [1,3,5,10,15,20,25]
A = [0.58278, 0.60264, 0.73509, 0.774834, 0.81456, 0.80794, 0.80794]# 50 n_estimators
B = [0.57615, 0.61589, 0.72847, 0.78145, 0.77483, 0.80132, 0.80132]# 100 n_estimators
C = [0.57615, 0.62251, 0.72847, 0.78807, 0.7947, 0.7947, 0.78145]# 200 n_estimators
D = [0.57615, 0.62251, 0.72847, 0.781456, 0.80794, 0.81456, 0.81456]# 300 n_estimators
E = [0.58278, 0.62251, 0.72847, 0.781456, 0.80794, 0.81456, 0.81456]# 400 n_estimators
F = [0.58278, 0.62251, 0.72847, 0.78807, 0.81456, 0.81456, 0.81456]# 500 n_estimators


plt.figure(figsize = (8,8))
plt.plot(max_depth, A, "*-", label = f"n_estimators: {n_estimators[0]}")
plt.plot(max_depth, B, "*-", label = f"n_estimators: {n_estimators[1]}")
plt.plot(max_depth, C, "*-", label = f"n_estimators: {n_estimators[2]}")
plt.plot(max_depth, D, "*-", label = f"n_estimators: {n_estimators[3]}")
plt.plot(max_depth, E, "*-", label = f"n_estimators: {n_estimators[4]}")
plt.plot(max_depth, F, "*-", label = f"n_estimators: {n_estimators[5]}")
plt.xlabel("Max Depth")
plt.ylabel("Validation Accuracy")
#plt.xscale("log")
plt.title("Validation Accuracy of Random Forest model with respect to max_depth and n_estimators")
plt.legend(loc = "best")
plt.grid(True)
plt.show()
```
