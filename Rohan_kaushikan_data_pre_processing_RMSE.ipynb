```python
source_loc = '/content/drive/Shareddrives/1:1_Rohan_Kaushikan/Dataset/Invasive Ductal Carcinoma/New Dataset'
label_column = 'Disease_Free_Status'
```

```python
import pandas as pd

data_location = source_loc + '/sorted_genes.csv'
data = pd.read_csv(data_location)

data
```

# **Clean the dataset**

```python
# Drop missing values
data = data.dropna()

# Drop duplicates
data.drop_duplicates(subset=None, keep="first", inplace=True)

data.head()
```

```python
# Get the no. of unique values of each class in the label column
data[label_column].value_counts()
```

```python
# Drop unwanted rows
data.drop(data.index[(data[label_column] == "[Not Available]")],axis=0,inplace=True)
```

```python
data
```

# **Train test split**

```python
# Split the dataset into train and test datasets
from sklearn.model_selection import train_test_split

# Label column
y = data[label_column]

# Feature columns
X = data.loc[:, data.columns != label_column]

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=42)

# Reset the index
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
X_val = X_val.reset_index(drop=True)
y_val = y_val.reset_index(drop=True)

# Display the shapes of the sets
print("Train set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
print("Val set shape:", X_val.shape, y_val.shape)
```

```python
y.shape
```

# **Scaling data**

```python
from sklearn import preprocessing
import pickle

# Normalize so everything is on the same scale.
cols = X_train.columns
min_max_scaler = preprocessing.MinMaxScaler().fit(X_train)
# Save the model
model_name = source_loc + "/model_min_max_scaler"
fh = open(model_name, "wb")
pickle.dump(min_max_scaler, fh)
fh.close()

# Scale train data
np_scaled_train = min_max_scaler.transform(X_train)
# New train data frame with the new scaled data.
X_train = pd.DataFrame(np_scaled_train, columns = cols)

# Scale test data
np_scaled_test = min_max_scaler.transform(X_test)
# New test data frame with the new scaled data.
X_test = pd.DataFrame(np_scaled_test, columns = cols)

# Scale val data
np_scaled_val = min_max_scaler.transform(X_val)
# New test data frame with the new scaled data.
X_val = pd.DataFrame(np_scaled_val, columns = cols)
```

# **Save train test datasets**

```python
# Train dataset
train_df = pd.concat([X_train, y_train], axis=1)
train_df.to_csv(source_loc + "/train.csv", index=False)

# Test dataset
test_df = pd.concat([X_test, y_test], axis=1)
test_df.to_csv(source_loc + "/test.csv", index=False)

# Val dataset
val_df = pd.concat([X_val, y_val], axis=1)
val_df.to_csv(source_loc + "/val.csv", index=False)
```
